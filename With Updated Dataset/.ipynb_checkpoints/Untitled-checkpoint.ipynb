# Using SMOTE dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
import sklearn.metrics as metrics
import operator
```

```python
X_test=pd.read_csv('X_test.csv')
X_train=pd.read_csv('X_train.csv')
y_test=pd.read_csv('y_test.csv')
y_train=pd.read_csv('y_train.csv')
```

```python
x_test = X_test.drop(['Gender','AgeM','ECV','TPP','SC','OAE1','OAE1.4','OAE2','OAE2.8','OAE4','Ear coded'], axis = 1)
x_train = X_train.drop(['Gender','AgeM','ECV','TPP','SC','OAE1','OAE1.4','OAE2','OAE2.8','OAE4','Ear coded'], axis = 1)
```

```python
index= x_test.columns.get_loc('fa8000')
x_train=x_train.iloc[:,:107]
x_test=x_test.iloc[:,:107]
```

```python
allModel = LogisticRegression(penalty="elasticnet",solver="saga",l1_ratio=0.5,class_weight='balanced')
allModel.fit(x_train, y_train)

importance = allModel.coef_[0]
importance = abs(importance)
```

# Coefficient results for all frequencies

```python
plt.figure(figsize=(20,5))
plt.plot(x_train.columns,importance)
plt.xticks(rotation=90)
plt.show()
```

# Top 10 coefficients

```python
freqKeys=list(x_train.columns)

d = {freqKeys[i]: importance[i] for i in range(len(freqKeys))}
sorted_d = dict( sorted(d.items(), key=operator.itemgetter(1),reverse=True))

dict_items = sorted_d.items()
top10 = list(dict_items)[:10]
print(top10)
```

```python
top10=dict(top10)
plt.bar(*zip(*top10.items()))
plt.xticks(rotation=90)
plt.show()
```
